```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
T, dt, R, B = 20, 0.1, 0.3, 1.2
N = int(T / dt)
time = np.linspace(0, T, N)
vel = np.ones(N)
ang_vel = np.zeros(N)
ang_vel[50:100] = 0.2
```

```python
left = vel - (ang_vel * B / 2) + np.random.normal(0, 0.02, N)
right = vel + (ang_vel * B / 2) + np.random.normal(0, 0.02, N)
distance = np.cumsum((left + right) / 2 * dt)
```

```python
x, y, theta = 0, 0, 0
pos = []
for i in range(N):
    theta += ang_vel[i] * dt
    x += vel[i] * np.cos(theta) * dt
    y += vel[i] * np.sin(theta) * dt
    pos.append((x, y))
pos = np.array(pos)
```

```python
plt.subplot(1, 2, 1)
plt.plot(time, distance, label="Distance")
plt.legend()
plt.subplot(1, 2, 2)
plt.plot(pos[:, 0], pos[:, 1], label="Path")
plt.legend()
plt.show()
```
